##Filling NuLL Values

```python
df2 = df.fillna(value=0)
df2
```

```python
df3 = df.fillna(value = 5)
df3
```

## Filling NuLL Values with a previous value

```python
df4 = df.fillna(method = 'pad')
df4
```

```python
df4.isnull().sum()
```

```python
## Filling NuLL Values with a previous value
df5 = df.fillna(method = 'bfill')
df5
```

```python
df6 = df.fillna(method = 'pad', axis = 1)
df6
```

```python
df7 = df.fillna(method = 'bfill', axis = 1)
df7
```

```python
df
```

```python
## Filling different Values in null in different columns
df8 = df.fillna({'society' : 'bcd', 'balcony':'def' })
df8
```

```python
## Filling null Value with the mean of columns
df9 = df.fillna(value = df['balcony'].mean())
df9
```

```python
df9 = df.fillna(value = df['balcony'].max())
df9
```

```python
df9 = df.fillna(value = df['balcony'].min())
df9
```

##Dropna

```python
df10 = df.dropna()
df10
```

```python
df11 = df.dropna(how = 'any')
df11
```

```python
df12 = df.dropna(how = 'all')
df12
```

##Replace

```python
df13 = df.replace(to_replace = np.nan, value = 'abcd123')
df13
```

```python
df13 = df.replace(to_replace = 3.0, value = '5.0')
df13
```

##Interpolate()

```python
df['balcony'] = df['balcony'].interpolate(method = 'linear')
df
```

```python
df['balcony'] = df['balcony'].interpolate(method = 'linear', limit_direction='forward')
df
```
